# Preprocessamento dos Dados

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
# Aplicando Label Encoder nas colunas categoricas
df['Month'] = label_encoder.fit_transform(df['Month'])
df['VisitorType'] = label_encoder.fit_transform(df['VisitorType'])
```

```python
#Convertendo "True" para 1 e "False" para 0 nas colunas "Revenue" e "Weekend
#df['Revenue'] = df['Revenue'].map({'True': 1, 'False': 0})
#df['Weekend'] = df['Weekend'].map({'True': 1, 'False': 0})
```

```python
#Escalonamento dos dados
from sklearn.preprocessing import StandardScaler
```

```python
df_scaler = StandardScaler()
```

# PrevisÃ£o de IntenÃ§Ã£o de Compra

```python
#divisÃ£o de previsores e classe
X = df.drop('Revenue', axis=1)
y = df['Revenue']
```

```python
X = df_scaler.fit_transform(X)
```

```python
#divisÃ£o treino e teste
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size= 0.30, random_state=0)
```

# Testando Modelo Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest_shopper = RandomForestClassifier(n_estimators= 10, criterion='entropy', random_state=0)
```

```python
random_forest_shopper.fit(X_treino, y_treino)
```

```python
previsao_shopper = random_forest_shopper.predict(X_teste)
```

```python
#gerando as metricas
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Avaliar o desempenho do modelo
acuracia = accuracy_score(y_teste, previsao_shopper)
```

```python
matriz_confusao = confusion_matrix(y_teste, previsao_shopper)
```

```python
relatorio_classificacao = classification_report(y_teste, previsao_shopper)
```

```python
print(f'AcurÃ¡cia: {acuracia:.4f}')
```

```python
print('Matriz de ConfusÃ£o:\n', matriz_confusao)
```

```python
print('RelatÃ³rio de ClassificaÃ§Ã£o:\n', relatorio_classificacao)
```

# Testando Modelo SVM

```python
from sklearn.svm import SVC
```

```python
svm_shopper = SVC(kernel='linear', random_state=1, C=1.0)
```

```python
svm_shopper.fit(X_treino, y_treino)
```

```python
previsao_svm = svm_shopper.predict(X_teste)
previsao_svm
```

```python
# Avaliar o desempenho do modelo
acuracia_svm = accuracy_score(y_teste, previsao_svm)
matriz_confusao_svm = confusion_matrix(y_teste, previsao_svm)
relatorio_classificacao_svm = classification_report(y_teste, previsao_svm)

print(f'AcurÃ¡cia: {acuracia_svm:.4f}')
print('Matriz de ConfusÃ£o:\n', matriz_confusao_svm)
print('RelatÃ³rio de ClassificaÃ§Ã£o:\n', relatorio_classificacao_svm)
```

# Testanto Modelo XGBoost

```python
import xgboost as xgb
```

```python
xgboost_shopper = xgb.XGBClassifier(objective='binary:logistic', random_state=42)
```

```python
# Treinar o modelo
xgboost_shopper.fit(X_treino, y_treino)
```

```python
# Fazer previsÃµes nos dados de teste
previsao_xgb = xgboost_shopper.predict(X_teste)
```

```python
# Avaliar o desempenho do modelo
acuracia_xgb = accuracy_score(y_teste, previsao_xgb)
matriz_confusao_xgb = confusion_matrix(y_teste, previsao_xgb)
relatorio_classificacao_xgb = classification_report(y_teste, previsao_xgb)

print(f'AcurÃ¡cia: {acuracia_xgb:.4f}')
print('Matriz de ConfusÃ£o:\n', matriz_confusao_xgb)
print('RelatÃ³rio de ClassificaÃ§Ã£o:\n', relatorio_classificacao_xgb)
```

# Escolhendo o Modelo

```python
# AcurÃ¡cia para cada modelo
acuracia_rf = accuracy_score(y_teste, previsao_shopper)
acuracia_svm = accuracy_score(y_teste, previsao_svm)
acuracia_xgboost = accuracy_score(y_teste, previsao_xgb)
acuracia_mlp = accuracy_score(y_teste, previsao_mlp)

print(f'AcurÃ¡cia Random Forest: {acuracia_rf:.4f}')
print(f'AcurÃ¡cia SVM: {acuracia_svm:.4f}')
print(f'AcurÃ¡cia XGBoost: {acuracia_xgboost:.4f}')
print(f'AcurÃ¡cia MLP: {acuracia_mlp:.4f}')
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
# Calcular probabilidades para cada modelo
y_prob_rf = random_forest_shopper.predict_proba(X_teste)[:, 1]
y_prob_svm = svm_shopper.decision_function(X_teste)
y_prob_xgboost = xgboost_shopper.predict_proba(X_teste)[:, 1]
y_prob_mlp = rede_neural_mlp.predict_proba(X_teste)[:, 1]
```

```python
# Calcular AUC para cada modelo
auc_rf = roc_auc_score(y_teste, y_prob_rf)
auc_svm = roc_auc_score(y_teste, y_prob_svm)
auc_xgboost = roc_auc_score(y_teste, y_prob_xgboost)
auc_mlp = roc_auc_score(y_teste, y_prob_mlp)
```

```python
# GrÃ¡fico das curvas ROC para cada modelo
plt.figure(figsize=(10, 8))
plt.plot(*roc_curve(y_teste, y_prob_rf)[:2], label=f'Random Forest (AUC = {auc_rf:.2f})')
plt.plot(*roc_curve(y_teste, y_prob_svm)[:2], label=f'SVM (AUC = {auc_svm:.2f})')
plt.plot(*roc_curve(y_teste, y_prob_xgboost)[:2], label=f'XGBoost (AUC = {auc_xgboost:.2f})')
plt.plot(*roc_curve(y_teste, y_prob_mlp)[:2], label=f'MLP (AUC = {auc_mlp:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='AleatÃ³rio')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curvas ROC para Modelos')
plt.legend()
plt.show()
```
